```python
box_pts = 9

def smooth_data(feats, labels):
    box = np.ones(box_pts)/box_pts
    
    smooth_feats = np.array(feats).copy()
    
#    DX = 3
#     DY = 4
#     AFX = 8
#     AFY = 9
#     AFZ = 10
    smooth_feats_idx = [3,4,5,6,7,8,9,10]
    
    for i in smooth_feats_idx:
        smooth_feats[:,i] = np.convolve(smooth_feats[:,i], box, mode='same')
    
    smooth_labels = np.array(labels).copy()
    _, label_num = smooth_labels.shape
    
    for i in range(label_num):
        smooth_labels[:,i] = np.convolve(smooth_labels[:,i], box, mode='same')
        
#     visualization(feats, labels, smooth_feats, smooth_labels)
    return smooth_feats, smooth_labels
        
```

```python
import numpy as np
import glob
import os


#rm first
SMOOTH_TRAIN_FILES = '/home/dataset/team2/training/training*/*_smooth_{}.csv'.format(box_pts)
SMOOTH_TEST_FILES = '/home/dataset/team2/validation/validation*/*_smooth_{}.csv'.format(box_pts)

smooth_train_csvs = glob.glob(SMOOTH_TRAIN_FILES)
smooth_test_csvs = glob.glob(SMOOTH_TEST_FILES)

for f in smooth_train_csvs:
    os.remove(f)

for f in smooth_test_csvs:
    os.remove(f)

TRAIN_FILES = '/home/dataset/team2/training/training*/*.csv'
TEST_FILES = '/home/dataset/team2/validation/validation*/*.csv'

train_csvs = glob.glob(TRAIN_FILES)
val_csvs = glob.glob(TEST_FILES)
print('train csv num:', len(train_csvs))
print('val csv num:', len(val_csvs))
```

```python
from tqdm.notebook import tqdm
import pandas as pd
import os

def write_to_csv(filename, feats, labels):
    comb_np = np.hstack((feats, labels))
    np.savetxt(filename, comb_np, delimiter=",")

train_num = len(train_csvs)
for i in tqdm(range(train_num)):
    file_path = train_csvs[i]
    raw = pd.read_csv(file_path).values
    feats = raw[:, :-3]
    labels = raw[:, -3:]
    smooth_feats, smooth_labels = smooth_data(feats, labels)
    smooth_path = file_path[:-4] + '_smooth_{}.csv'.format(box_pts)
    
#     print(smooth_path)
    write_to_csv(smooth_path, smooth_feats, smooth_labels)
    
```

```python
val_num = len(val_csvs)
for i in tqdm(range(val_num)):
    file_path = val_csvs[i]
    raw = pd.read_csv(file_path).values
    feats = raw[:, :-3]
    labels = raw[:, -3:]
    smooth_feats, smooth_labels = smooth_data(feats, labels)
    smooth_path = file_path[:-4] + '_smooth_{}.csv'.format(box_pts)
    
#     print(smooth_path)
    write_to_csv(smooth_path, smooth_feats, smooth_labels)
    
#     break

# find path
# file_path = train_csvs[44]
# name_str = (file_path.split('/')[-1])
# tar_name = '_'.join(name_str.split('_')[:2])
# file_name = '_'.join(name_str.split('_')[2:])
# file_name = file_name[:-4]
# folder_path = '/home/dataset/{}/{}/'.format(tar_name, file_name)
```
